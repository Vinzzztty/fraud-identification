```python
import numpy as np
import pandas as pd

from Extract_Features.extract_features_vibration import (
    bearing_fault_frequencies,
    fft
)

# =========================
# BASELINE (SUDAH ADA)
# =========================
baseline = {
    "bpfo": bpfo_base,
    "bpfi": bpfi_base,
    "bsf": bsf_base,
    "ftf": ftf_base,
    "amp_2x": amp2x_base,
    "rms": rms_base,
    "kurtosis": kurt_base
}

# =========================
# WORST (SUDAH ADA)
# =========================
worst = {
    "bpfo": bpfo_worst,
    "bpfi": bpfi_worst,
    "bsf": bsf_worst,
    "ftf": ftf_worst,
    "amp_2x": amp2x_worst,
    "rms": rms_worst,
    "kurtosis": kurt_worst
}


# =========================
# RAW FFT (for 1X, 2X)
# =========================
def fft_raw(signal, sampling_rate):
    n = len(signal)
    window = np.hanning(n)
    sig = signal * window

    fft_vals = np.fft.rfft(sig)
    mag = np.abs(fft_vals) * (2.0 / n)
    freq = np.fft.rfftfreq(n, 1/sampling_rate)

    return freq, mag


# =========================
# AMPLITUDE PICK
# =========================
def amplitude_near(freq, values, target_freq):
    idx = np.argmin(np.abs(freq - target_freq))
    return values[idx]


# =========================
# FEATURE EXTRACTION
# =========================
def extract_raw_features(signal, rpm, sampling_rate):

    n_ball = 8
    Pd = 72.5
    Bd = 15

    # RAW FFT
    freq_raw, mag_raw = fft_raw(signal, sampling_rate)
    f_1x = rpm / 60

    amp_2x = amplitude_near(freq_raw, mag_raw, 2 * f_1x)

    # ENVELOPE FFT
    freq_env, _, mag_env = fft(signal, sampling_rate)
    BPFO, BPFI, BSF, FTF = bearing_fault_frequencies(rpm, n_ball, Bd, Pd)

    amp_bpfo = amplitude_near(freq_env, mag_env, BPFO)
    amp_bpfi = amplitude_near(freq_env, mag_env, BPFI)
    amp_bsf  = amplitude_near(freq_env, mag_env, BSF)
    amp_ftf  = amplitude_near(freq_env, mag_env, FTF)

    # TIME DOMAIN
    rms = np.sqrt(np.mean(signal**2))
    kurt = pd.Series(signal).kurtosis()

    return {
        "bpfo": amp_bpfo,
        "bpfi": amp_bpfi,
        "bsf": amp_bsf,
        "ftf": amp_ftf,
        "amp_2x": amp_2x,
        "rms": rms,
        "kurtosis": kurt
    }


# =========================
# NORMALIZATION
# =========================
def normalize(value, base, worst):
    return (value - base) / (worst - base + 1e-10)


# =========================
# HITUNG HI
# =========================
def compute_hi(features, baseline, worst):

    # bearing energy
    bearing_val = features["bpfo"] + features["bpfi"] + features["bsf"] + features["ftf"]

    #bearing normalize
    bpfo_HI = abs((features["bpfo"] - baseline["bpfo"]) / (worst["bpfo"] - baseline["bpfo"]))
    bpfi_HI = abs((features ["bpfi"] - baseline["bpfi"]) / (worst["bpfi"] - baseline["bpfi"]))
    bsf_HI = abs((features["bsf"] - baseline["bsf"]) / (worst["bsf"] - baseline["bsf"]))
    bsf_HI = abs((features["ftf"] - baseline["ftf"]) / (worst["ftf"] - baseline["ftf"]))
    
    # bearing energy
    bearing_val = features["bpfo"] + features["bpfi"] + features["bsf"] + features["ftf"]
    bearing_base = baseline["bpfo"] + baseline["bpfi"] + baseline["bsf"] + baseline["ftf"]
    bearing_worst = worst["bpfo"] + worst["bpfi"] + worst["bsf"] + worst["ftf"]

    s_bpfo = normalize(features["bpfo"], baseline["bpfo"], worst["bpfo"])
    s_bpfi = normalize(features["bpfi"], baseline["bpfi"], worst["bpfi"])
    s_bsf = normalize(features["bsf"], baseline["bsf"], worst["bsf"])
    s_ftf = normalize(features["ftf"], baseline["ftf"], worst["ftf"])
    s_bearing = max(s_bpfo, s_bpfi, s_bsf, s_ftf)
    
    # s_bearing = normalize(bearing_val, bearing_base, bearing_worst)
    s_harm = normalize(features["amp_2x"], baseline["amp_2x"], worst["amp_2x"])
    s_rms = normalize(features["rms"], baseline["rms"], worst["rms"])
    s_kurt = normalize(features["kurtosis"], baseline["kurtosis"], worst["kurtosis"])

    # clip 0-1
    s_bearing = np.clip(s_bearing, 0, 1)
    s_harm = np.clip(s_harm, 0, 1)
    s_rms = np.clip(s_rms, 0, 1)
    s_kurt = np.clip(s_kurt, 0, 1)

    hi = 1 - (
        0.6 * s_bearing +
        0.2 * s_harm +
        0.2 * s_rms +
        0 * s_kurt
    )

    return np.clip(hi, 0, 1)


# =========================
# EVALUATE FILE
# =========================
def evaluate_dataset(file_path, rpm, sampling_rate, baseline, worst):

    df = pd.read_csv(file_path)
    signal_columns = df.columns[1:]

    hi_list = []

    for col in signal_columns:
        signal = df[col].values

        features = extract_raw_features(signal, rpm, sampling_rate)
        hi = compute_hi(features, baseline, worst)

        hi_list.append(hi)

    return hi_list


# =========================
# ===== MAIN TEST =====
# =========================

file1 = file_path_new_50
file2 = Rtf_1
file3 = Rtf_65
file4 = Rtf_128

rpm = 1470 * 0.5
sampling_rate = 12000

results = {}

for i, file in enumerate([file1, file2, file3, file4], start=1):

    hi_values = evaluate_dataset(file, rpm, sampling_rate, baseline, worst)

    results[f"file_{i}"] = {
        "mean": float(np.mean(hi_values)),
        "min": float(np.min(hi_values)),
        "max": float(np.max(hi_values))
    }


# =========================
# PRINT RESULT
# =========================
print("\n=== HI RESULTS ===")

for name, res in results.items():
    print(f"\n{name.upper()}")
    print("Mean:", res["mean"])
    print("Min :", res["min"])
    print("Max :", res["max"])
```

```python
import numpy as np
import pandas as pd

from Extract_Features.extract_features_vibration import (
    bearing_fault_frequencies,
    fft
)

# =========================
# BASELINE & WORST (SUDAH ADA)
# =========================
baseline = {
    "bpfo": bpfo_base,
    "bpfi": bpfi_base,
    "bsf": bsf_base,
    "ftf": ftf_base,
    "amp_2x": amp2x_base,
    "rms": rms_base
}

worst = {
    "bpfo": bpfo_worst,
    "bpfi": bpfi_worst,
    "bsf": bsf_worst,
    "ftf": ftf_worst,
    "amp_2x": amp2x_worst,
    "rms": rms_worst
}


# =========================
# AMPLITUDE (1 TITIK)
# =========================
def amplitude_near(freq, values, target_freq):
    idx = np.argmin(np.abs(freq - target_freq))
    return values[idx]


# =========================
# FEATURE EXTRACTION (FIXED)
# =========================
def extract_raw_features(signal, rpm, sampling_rate):

    n_ball = 8
    Pd = 72.5
    Bd = 15

    # =========================
    # ENVELOPE FFT
    # =========================
    freq_env, _, mag_env = fft(signal, sampling_rate)

    # 🔥 CONVERT TO dB
    db_env = 20 * np.log10(mag_env + 1e-12)

    BPFO, BPFI, BSF, FTF = bearing_fault_frequencies(rpm, n_ball, Bd, Pd)

    amp_bpfo = amplitude_near(freq_env, db_env, BPFO)
    amp_bpfi = amplitude_near(freq_env, db_env, BPFI)
    amp_bsf  = amplitude_near(freq_env, db_env, BSF)
    amp_ftf  = amplitude_near(freq_env, db_env, FTF)

    # =========================
    # RAW FFT → 2X harmonic
    # =========================
    n = len(signal)
    window = np.hanning(n)
    sig = signal * window

    fft_vals = np.fft.rfft(sig)
    mag = np.abs(fft_vals)

    # 🔥 CONVERT TO dB
    db = 20 * np.log10(mag + 1e-12)
    freq = np.fft.rfftfreq(n, 1/sampling_rate)

    f_1x = rpm / 60
    amp_2x = amplitude_near(freq, db, 2 * f_1x)

    # =========================
    # TIME DOMAIN
    # =========================
    rms = np.sqrt(np.mean(signal**2))

    return {
        "bpfo": amp_bpfo,
        "bpfi": amp_bpfi,
        "bsf": amp_bsf,
        "ftf": amp_ftf,
        "amp_2x": amp_2x,
        "rms": rms
    }


# =========================
# NORMALIZATION
# =========================
def normalize(value, base, worst):
    return (value - base) / (worst - base + 1e-10)


# =========================
# HITUNG HI
# =========================
def compute_hi(features, baseline, worst):

    s_bpfo = normalize(features["bpfo"], baseline["bpfo"], worst["bpfo"])
    s_bpfi = normalize(features["bpfi"], baseline["bpfi"], worst["bpfi"])
    s_bsf  = normalize(features["bsf"],  baseline["bsf"],  worst["bsf"])
    s_ftf  = normalize(features["ftf"],  baseline["ftf"],  worst["ftf"])

    # 🔥 ambil yang paling dominan
    s_bearing = max(s_bpfo, s_bpfi, s_bsf, s_ftf)

    s_harm = normalize(features["amp_2x"], baseline["amp_2x"], worst["amp_2x"])
    s_rms  = normalize(features["rms"], baseline["rms"], worst["rms"])

    # clip
    s_bearing = np.clip(s_bearing, 0, 1)
    s_harm = np.clip(s_harm, 0, 1)
    s_rms = np.clip(s_rms, 0, 1)

    hi = 1 - (
        0.6 * s_bearing +
        # 0.25 * s_harm +
        0.4 * s_rms
    )

    return np.clip(hi, 0, 1)


# =========================
# EVALUATE FILE
# =========================
def evaluate_dataset(file_path, rpm, sampling_rate):

    df = pd.read_csv(file_path)
    signal_columns = df.columns[1:]

    hi_list = []

    for col in signal_columns:
        signal = df[col].values

        features = extract_raw_features(signal, rpm, sampling_rate)
        hi = compute_hi(features, baseline, worst)

        hi_list.append(hi)

    return hi_list


# =========================
# ===== MAIN TEST =====
# =========================

file1 = file_path_new_50
file2 = file_path_bpfo1_50
file3 = file_path_bpfo2_50
file4 = file_path_bpfo3_50

rpm = 1470 * 0.5
sampling_rate = 12000

results = {}

for i, file in enumerate([file1, file2, file3, file4], start=1):

    hi_values = evaluate_dataset(file, rpm, sampling_rate)

    results[f"file_{i}"] = {
        "mean": float(np.mean(hi_values)),
        "min": float(np.min(hi_values)),
        "max": float(np.max(hi_values))
    }


# =========================
# PRINT RESULT
# =========================
print("\n=== HI RESULTS ===")

for name, res in results.items():
    print(f"\n{name.upper()}")
    print("Mean:", res["mean"])
    print("Min :", res["min"])
    print("Max :", res["max"])
```

```python
for i, file in enumerate([file1, file2, file3, file4], start=1):

    print(f"\n========== FILE {i} ==========")

    # =========================
    # HITUNG HI
    # =========================
    hi_values = evaluate_dataset(file, rpm, sampling_rate)

    print("HI Mean :", np.mean(hi_values))
    print("HI Min  :", np.min(hi_values))
    print("HI Max  :", np.max(hi_values))

    # =========================
    # EXTRACT ALL FEATURES
    # =========================
    df = pd.read_csv(file)
    signal_columns = df.columns[1:]

    all_features = []

    for col in signal_columns:
        signal = df[col].values
        feat = extract_raw_features(signal, rpm, sampling_rate)
        all_features.append(feat)

    df_feat = pd.DataFrame(all_features)

    # =========================
    # FEATURE dB (SUDAH ADA SEBENARNYA)
    # =========================
    print("\n--- FEATURE dB (MEAN) ---")
    print(df_feat.mean())

    print("\n--- FEATURE dB (MIN) ---")
    print(df_feat.min())

    print("\n--- FEATURE dB (MAX) ---")
    print(df_feat.max())


    # =========================
    # HITUNG SEVERITY (HI COMPONENT)
    # =========================
    def compute_severity(df_feat, baseline, worst):

        def norm(val, base, w):
            return (val - base) / (w - base + 1e-10)

        results = {}

        for mode in ["mean", "min", "max"]:

            if mode == "mean":
                f = df_feat.mean()
            elif mode == "min":
                f = df_feat.min()
            else:
                f = df_feat.max()

            # =========================
            # BEARING
            # =========================
            s_bpfo = norm(f["bpfo"], baseline["bpfo"], worst["bpfo"])
            s_bpfi = norm(f["bpfi"], baseline["bpfi"], worst["bpfi"])
            s_bsf  = norm(f["bsf"],  baseline["bsf"],  worst["bsf"])
            s_ftf  = norm(f["ftf"],  baseline["ftf"],  worst["ftf"])

            # 🔥 sesuai kode kamu
            s_bearing = min(s_bpfo, s_bpfi, s_bsf, s_ftf)

            # =========================
            # OTHERS
            # =========================
            s_harm = norm(f["amp_2x"], baseline["amp_2x"], worst["amp_2x"])
            s_rms  = norm(f["rms"], baseline["rms"], worst["rms"])

            # clip
            s_bearing = np.clip(s_bearing, 0, 1)
            s_harm = np.clip(s_harm, 0, 1)
            s_rms = np.clip(s_rms, 0, 1)

            # =========================
            # HI FINAL
            # =========================
            hi = 1 - (0.6 * s_bearing + 0.25 * s_harm + 0.15 * s_rms)

            results[mode] = {
                "s_bpfo": float(s_bpfo),
                "s_bpfi": float(s_bpfi),
                "s_bsf": float(s_bsf),
                "s_ftf": float(s_ftf),
                "s_bearing": float(s_bearing),
                "s_harm": float(s_harm),
                "s_rms": float(s_rms),
                "HI": float(np.clip(hi, 0, 1))
            }

        return results


    # =========================
    # PRINT HASIL
    # =========================
    severity = compute_severity(df_feat, baseline, worst)

    print("\n--- HI COMPONENT & FINAL HI ---")

    for mode, val in severity.items():

        print(f"\n[{mode.upper()}]")

        print("s_bpfo    :", f"{val['s_bpfo']:.3f}")
        print("s_bpfi    :", f"{val['s_bpfi']:.3f}")
        print("s_bsf     :", f"{val['s_bsf']:.3f}")
        print("s_ftf     :", f"{val['s_ftf']:.3f}")

        print("s_bearing :", f"{val['s_bearing']:.3f}")
        print("s_harm    :", f"{val['s_harm']:.3f}")
        print("s_rms     :", f"{val['s_rms']:.3f}")

        print("HI        :", f"{val['HI']:.3f}")
```
